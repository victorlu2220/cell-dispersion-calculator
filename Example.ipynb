```python
import matplotlib.pyplot as plt
from matplotlib.image import imread
import numpy as np
from scipy.spatial import ConvexHull
import time
```

```python
# import data
image_data = imread("coverlip Day 7.png")

# display image and configure size of display window
plt.rcParams['figure.figsize'] = [40, 30]    
plt.imshow(image_data)
plt.show()

# initialize total area and cell area
total_area = 0
cell_area = 0
nuclei_area = 0

for row in range(len(image_data)):
    for column in range(len(image_data[row])):
        r = image_data[row][column][0]
        g = image_data[row][column][1]
        b = image_data[row][column][2]
        
        if r <= 100/255 and g <= 100/255 and b <= 100/255: # check for black spots, turn them to white
            image_data[row][column][0] = 1
            image_data[row][column][1] = 1
            image_data[row][column][2] = 1
        else: # if not black then it contains cells
            if b >= 110/255 and r >= 110/255: 
                image_data[row][column][0] = 0
                image_data[row][column][1] = 0
                image_data[row][column][2] = 1
                nuclei_area += 1
            cell_area += 1
        total_area += 1

# Show processed image
plt.imshow(image_data)
cell_per_area = cell_area/total_area
print(cell_per_area)

    
```
